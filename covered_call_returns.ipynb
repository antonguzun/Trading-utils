# Covered Call Return Calculator
This notebook computes returns for covered call options using formulas from *Option Strategies*. Parameters are defined below and can be adjusted to explore different scenarios.

```
# Initial parameters
shares = 500
stock_price = 43.0
strike_price = 45.0
option_price = 3.0  # premium received per share
dividends = 500.0  # total dividends during option life
comm_stock_buy = 320.0
comm_stock_sell = 330.0
comm_option_sell = 60.0
margin_rate = 0.50  # fraction of purchase paid with investor equity
interest_rate = 0.10  # annual interest on borrowed funds
months = 6
```

```
from covered_call import calc_cash_returns, calc_margin_returns
```

```
cash = calc_cash_returns(shares, stock_price, strike_price, option_price, dividends,
                         comm_stock_buy, comm_stock_sell, comm_option_sell)
margin = calc_margin_returns(shares, stock_price, strike_price, option_price, dividends,
                             comm_stock_buy, comm_stock_sell, comm_option_sell,
                             margin_rate, interest_rate, months)

print('Cash account results:')
print(cash)
print()
print('Margin account results:')
print(margin)
```
